## 1b. Kiến trúc Star Schema — 3 Fact + 4 Dim

**Slide note:** Từ fact phẳng `fact_real_estate.csv` (284 quan sát), dữ liệu được chuẩn hóa thành **star schema** chuẩn kho dữ liệu: 3 bảng fact tách theo chủ đề nghiệp vụ + 4 bảng dimension dùng chung. Bảo toàn 279/279 quan sát hợp lệ, referential integrity PASS 100%.

```
                    ┌─────────────┐
                    │  dim_date   │ 21 kỳ (2024Q1→2026Q1, grain QUÝ + năm)
                    └──────┬──────┘
 ┌────────────┐            │            ┌─────────────┐
 │ dim_region ├───┐        │        ┌───┤ dim_segment │ 14 phân khúc, 5 nhóm
 │ 15 vùng,   │   │  ┌─────┴─────┐  │   └─────────────┘
 │ 3 cấp      │   ├──┤   FACT    ├──┤
 └────────────┘   │  │ supply /  │  │   ┌─────────────┐
                  │  │ demand /  │  ├───┤ dim_report  │ 12 báo cáo nguồn (lineage)
                  │  │ price     │  │   └─────────────┘
                  │  └───────────┘  │
                  └─────────────────┘
```

| Bảng | Rows | Grain / Nội dung | Measures chính |
|---|---|---|---|
| **`fact_supply`** | 113 | kỳ × vùng × phân khúc | `new_supply`, `da_hoan_thanh`, `du_dieu_kien_mo_ban`, `products_under_development__{can,lo,san_pham}` |
| **`fact_demand`** | 61 | kỳ × vùng × phân khúc | `transactions`, `absorption_rate`, `inventory` |
| **`fact_price`** | 54 | kỳ × vùng × phân khúc | `price_min/max`, `price_primary_avg__{trieu_dong_m2,usd_m2}`, `price_secondary_avg`, `rental_yield` |
| `dim_date` | 21 | 1 dòng/kỳ — `period_type` ∈ {quarter, year} (grain tự nhiên là **QUÝ** — Bộ Xây dựng chỉ công bố quý) | — |
| `dim_region` | 15 | 3 cấp: country / macro_region / province + `parent_region` để roll-up (tránh double-count khi SUM) | — |
| `dim_segment` | 14 | 14 phân khúc gom 5 nhóm `segment_group` (căn hộ / thấp tầng / đất nền / tổng hợp...) | — |
| `dim_report` | 12 | lineage: mỗi dòng fact trace ngược về báo cáo gốc qua cột `sources` | — |

**Join keys:** fact ↔ `dim_date` qua `date_key` · fact ↔ `dim_region` qua `region` · fact ↔ `dim_segment` qua `segment` · cột `sources` của fact ↔ `dim_report.report_id`.

**2 cột audit trên mọi fact:** `confidence` (high/medium/low — worst-of-group khi gộp) + `sources` (lineage báo cáo gốc).

````python
# 1b-i. Load star schema + kiểm tra referential integrity
dims = {n: pd.read_csv(f'{DS}/{n}.csv') for n in ['dim_date','dim_region','dim_segment','dim_report']}
facts = {n: pd.read_csv(f'{DS}/{n}.csv') for n in ['fact_supply','fact_demand','fact_price']}

print(f"{'Bảng':<14} {'Rows':>5}  Cột")
for n, df in {**dims, **facts}.items():
    print(f"{n:<14} {len(df):>5}  {df.shape[1]} cột")

# RI check: mọi khóa ngoại trong fact phải tồn tại trong dim
print('\nReferential integrity:')
for n, f in facts.items():
    ok_d = f['date_key'].isin(dims['dim_date']['date_key']).all()
    ok_r = f['region'].isin(dims['dim_region']['region']).all()
    ok_s = f['segment'].isin(dims['dim_segment']['segment']).all()
    print(f"  {n:<13} date_key:{'PASS' if ok_d else 'FAIL'}  region:{'PASS' if ok_r else 'FAIL'}  segment:{'PASS' if ok_s else 'FAIL'}")

key_cols = ['date_key','period','period_type','year','quarter','month','region','segment','confidence','sources']
n_rows = sum(len(f) for f in facts.values())
n_obs = sum(int(f.drop(columns=key_cols).notna().sum().sum()) for f in facts.values())
print(f'\nTổng: {n_rows} dòng fact chứa {n_obs} giá trị quan sát (1 dòng = 1 kỳ×vùng×phân khúc, nhiều measure)')
````

````python
# 1b-ii. Data dictionary các dimension
print('── dim_region: 3 cấp vùng (roll-up qua parent_region) ──')
print(dims['dim_region'].sort_values(['region_level','region']).to_string(index=False))
print('\n── dim_segment: 14 phân khúc / 5 nhóm ──')
print(dims['dim_segment'].sort_values(['segment_group','segment']).to_string(index=False))
print('\n── dim_date: phân bố kỳ ──')
print(dims['dim_date']['period_type'].value_counts().to_string())
print(f"  range: {dims['dim_date']['period'].min()} → {dims['dim_date']['period'].max()}")
````

````python
# 1b-iii. Ví dụ join star schema: trả lời câu hỏi đa chiều bằng 1 lệnh merge + pivot
# Q: Giao dịch theo NHÓM phân khúc × NĂM, chỉ cấp quốc gia (tránh double-count nhờ dim_region.region_level)
ex = (facts['fact_demand']
      .merge(dims['dim_region'], on='region')
      .merge(dims['dim_segment'], on='segment'))
nat = ex[(ex['region_level']=='country') & (ex['period_type']=='quarter')]
pv = nat.pivot_table(values='transactions', index='segment_group', columns='year', aggfunc=lambda s: s.sum(min_count=1))
print('Giao dịch cấp quốc gia (cộng các quý) — segment_group × năm:')
pv = pv.dropna(how='all').dropna(axis=1, how='all')
print(pv.applymap(lambda v: f'{v:,.0f}' if pd.notna(v) else '—').to_string())  # — = nguồn không công bố, KHÔNG phải 0

# Q2: giá sơ cấp TB theo cấp vùng (dim_region.region_level)
ex2 = facts['fact_price'].merge(dims['dim_region'], on='region').merge(dims['dim_segment'], on='segment')
pv2 = ex2.pivot_table(values='price_primary_avg__trieu_dong_m2', index='segment_group', columns='region_level', aggfunc='mean').round(1)
print('\nGiá sơ cấp TB (triệu đồng/m²) — segment_group × region_level:')
print(pv2.to_string())
print('\n→ dim_region.region_level giúp lọc đúng cấp tổng hợp, tránh double-count khi SUM — giá trị của star schema so với fact phẳng.')
````

## 2. Nguồn cung nhà ở toàn quốc (theo quý — Bộ Xây dựng)
**Slide note:** Sản phẩm đang triển khai ~400 nghìn; nguồn "đủ điều kiện mở bán" và "hoàn thành" phản ánh cung thực tế ra thị trường từng quý.

````python
sup = pd.read_csv(f'{DS}/re_supply_quarterly.csv')
piv = sup.pivot_table(index='quarter', columns='metric', values='value', aggfunc='sum')
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))
small = [c for c in ['du_dieu_kien_mo_ban','da_hoan_thanh','new_supply'] if c in piv.columns]
piv[small].plot(ax=axes[0], marker='o')
axes[0].set_title('Cung thực tế ra thị trường theo quý (căn)')
axes[0].set_ylabel('căn'); axes[0].grid(alpha=.3); axes[0].legend(loc='best', fontsize=9)
if 'products_under_development' in piv.columns:
    piv['products_under_development'].dropna().plot(ax=axes[1], marker='s', color='tab:red')
    axes[1].set_title('Sản phẩm đang triển khai (tồn pipeline)')
    axes[1].set_ylabel('sản phẩm'); axes[1].grid(alpha=.3)
for ax in axes: ax.tick_params(axis='x', rotation=45)
plt.tight_layout(); plt.show()
print(piv.round(0).to_string())
````

## 3. Nguồn cung mới theo thành phố (HN / ĐN / TP.HCM, 2022Q1→2025Q3)
**Slide note:** Chuỗi 15 quý cross-validated 100% giữa 2 báo cáo quý (08/2025 & 11/2025). Chung cư HN dẫn dắt chu kỳ phục hồi 2024-2025; TP.HCM phục hồi chậm hơn.

````python
city = pd.read_csv(f'{DS}/re_supply_city_quarterly.csv')
cq = city[city['granularity']=='quarter'].copy()
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5), sharex=True)
for cname, g in cq.groupby('city'):
    g = g.sort_values('period')
    axes[0].plot(g['period'], g['chung_cu'], marker='o', label=cname)
    axes[1].plot(g['period'], g['bds_gan_lien_dat'], marker='s', label=cname)
axes[0].set_title('Nguồn cung mới — Chung cư (căn/quý)')
axes[1].set_title('Nguồn cung mới — BĐS gắn liền đất (căn/quý)')
for ax in axes:
    ax.legend(fontsize=9); ax.grid(alpha=.3); ax.tick_params(axis='x', rotation=60)
plt.tight_layout(); plt.show()
# QoQ growth chuyên sâu
hn = cq[cq.city=='Hà Nội'].sort_values('period')
print('QoQ tăng trưởng cung chung cư Hà Nội (%):')
print((hn.set_index('period')['chung_cu'].pct_change()*100).round(1).dropna().to_string())
````

## 7. Phân tích chuyên sâu — chỉ số phái sinh
**Slide note (2 slide):**
- **Sức hấp thụ**: so giao dịch chung cư + nhà riêng lẻ với nguồn "đủ điều kiện mở bán" cùng quý — đo cầu thực so cung mới.
- **Cơ cấu TP.HCM mới sau sáp nhập** (Q3/2025): Bình Dương cũ đóng góp ~75% nguồn cung căn hộ vùng TP.HCM mở rộng.

````python
# 7a. Sức hấp thụ: transactions (chung cư + nhà riêng lẻ) / du_dieu_kien_mo_ban
mo_ban = sup[sup.metric=='du_dieu_kien_mo_ban'].set_index('quarter')['value']
gd_cc = tq[tq.segment.str.contains('chung_cu', na=False)].groupby('quarter')['transactions'].sum()
common = mo_ban.index.intersection(gd_cc.index)
absorb = pd.DataFrame({'cung_mo_ban': mo_ban[common], 'giao_dich_cc_nha': gd_cc[common]})
absorb['ty_le_hap_thu'] = (absorb['giao_dich_cc_nha']/absorb['cung_mo_ban']).round(2)
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))
absorb[['cung_mo_ban','giao_dich_cc_nha']].plot(kind='bar', ax=axes[0], width=.75)
axes[0].set_title('Cung đủ ĐK mở bán vs Giao dịch CC+nhà riêng lẻ')
axes[0].legend(['Cung mở bán','Giao dịch'], fontsize=9); axes[0].grid(alpha=.3, axis='y')
axes[0].tick_params(axis='x', rotation=45)
# 7b. Cơ cấu TP.HCM mới
hcm = pd.read_csv(f'{DS}/re_supply_hcm_region_2025q3.csv')
vals = hcm.set_index('region')['can_ho_chung_cu']
vals = vals[vals > 0]
axes[1].pie(vals, labels=vals.index, autopct='%1.0f%%', startangle=90,
            colors=['tab:blue','tab:orange','tab:green'][:len(vals)])
axes[1].set_title('Cơ cấu cung căn hộ TP.HCM mới — Q3/2025 (sau sáp nhập)')
plt.tight_layout(); plt.show()
print('Tỷ lệ hấp thụ (giao dịch / cung mở bán) theo quý:')
print(absorb['ty_le_hap_thu'].to_string())
````
